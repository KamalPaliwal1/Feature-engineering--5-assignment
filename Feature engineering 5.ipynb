✅ Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose one over the other.
1. Ordinal Encoding:
Definition: It assigns a unique integer to each category while maintaining the order of the categories.
When to Use: When the categories have a natural order (e.g., education levels or customer satisfaction).
Example:
python
Copy
Edit
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

data = np.array([['Low'], ['Medium'], ['High'], ['Medium'], ['Low']])
encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
encoded_data = encoder.fit_transform(data)
print(encoded_data)
Output:

css
Copy
Edit
[[0.]
 [1.]
 [2.]
 [1.]
 [0.]]
2. Label Encoding:
Definition: Assigns an integer to each unique category without considering order.
When to Use: When the categories are nominal (no order), such as gender or country names.
Example:
python
Copy
Edit
from sklearn.preprocessing import LabelEncoder

data = ['Red', 'Green', 'Blue', 'Green', 'Red']
encoder = LabelEncoder()
encoded_data = encoder.fit_transform(data)
print(encoded_data)
Output:

csharp
Copy
Edit
[2 1 0 1 2]
⚡️ When to Choose One Over the Other:
Ordinal Encoding: When the feature has a ranking or hierarchy.
Example: Satisfaction level → Low, Medium, High
Label Encoding: When the feature has no natural order.
Example: Gender → Male, Female






✅ Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.


Target Guided Ordinal Encoding:
Definition: Assigns ordinal values to categories based on their relationship with the target variable.
Steps:
Group the categorical feature by the target variable and calculate the mean of the target.
Sort the categories based on the mean target value.
Assign ranks to the categories based on their mean.
Example:
Consider a dataset where we predict whether a customer buys a product (1 or 0) based on the feature City.

City	Purchase
A	1
B	0
C	1
A	0
B	0
C	1
Steps:

Group by and calculate mean:
A → (1+0) / 2 = 0.5
B → (0+0) / 2 = 0.0
C → (1+1) / 2 = 1.0
Assign Ranks:
Rank based on target mean → B (0), A (1), C (2)
Encoded Data:
A → 1
B → 0
C → 2




✅ Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?
Covariance:
Definition: Covariance measures the relationship between two variables. It shows how one variable changes with respect to another.
hy Covariance is Important:
Positive Covariance: Variables move in the same direction.
Negative Covariance: Variables move in opposite directions.
Zero Covariance: No relationship between the variables.





✅ Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium, large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library. Show your code and explain the output.


from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Creating a dataset
data = {'Color': ['red', 'green', 'blue', 'green', 'red'],
        'Size': ['small', 'medium', 'large', 'small', 'medium'],
        'Material': ['wood', 'metal', 'plastic', 'wood', 'metal']}

df = pd.DataFrame(data)

# Applying Label Encoding
label_encoders = {}
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

print(df)

✅ Q5. Calculate the covariance matrix for the following variables in a dataset: Age, Income, and Education level. Interpret the results.

import numpy as np
import pandas as pd

# Creating a dataset
data = {'Age': [25, 35, 45, 20, 30],
        'Income': [50, 80, 90, 45, 70],
        'Education': [2, 3, 4, 1, 2]}

df = pd.DataFrame(data)

# Covariance Matrix
cov_matrix = df.cov()
print(cov_matrix)

✅ Q6. Encoding Methods for Gender, Education Level, and Employment Status
Gender (Male/Female): Use Label Encoding or One-Hot Encoding. One-hot encoding is preferred for models that need categorical values as binary vectors.
Education Level (High School, Bachelor's, Master's, PhD): Use Ordinal Encoding due to the natural order of educational levels.
Employment Status (Unemployed, Part-Time, Full-Time): Use One-Hot Encoding since it’s a nominal variable with no order.





✅ Q7. Covariance Calculation for Continuous and Categorical Variables
To calculate covariance between:

Temperature and Humidity: Standard covariance using np.cov().
Weather Condition and Wind Direction: Convert categorical variables using label encoding or one-hot encoding and then compute covariance.

import numpy as np
import pandas as pd

# Creating a sample dataset
data = {'Temperature': [30, 28, 35, 33, 29],
        'Humidity': [70, 65, 80, 75, 68],
        'Weather': ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Cloudy'],
        'Wind': ['North', 'South', 'East', 'West', 'North']}

df = pd.DataFrame(data)

# Encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Weather'] = le.fit_transform(df['Weather'])
df['Wind'] = le.fit_transform(df['Wind'])

# Covariance Matrix
cov_matrix = df.cov()
print(cov_matrix)
